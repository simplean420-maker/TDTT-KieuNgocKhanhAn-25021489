```python
c = str(input('Nhập một chữ cái: '))
if len(c) != 1 or not c.isalpha():
    print('Nhập một chữ cái')
else:
    if c.islower():
        a = c.upper()
        print(a)
    else:
        b = c.lower()
        print(b)
```

```python
import math
num1, num2, num3 = map(int, [input(f'Nhập cạnh thứ {i}: ') for i in range(1, 4)])
if num1 + num2 > num3 and num1 + num3 > num2 and num2 + num3 > num1:
    s = (num1 + num2 + num3) / 2 
    area = math.sqrt(s * (s - num1) * (s - num2) * (s - num3))
    print('Diện tích tam giác là: ', area)
else:
    print('Không phải 3 cạnh của tam giác')
```

```python
name = str(input('Nhập họ và tên chủ hộ: '))
trc = float(input('Nhập số điện tháng trước: '))
sau = float(input('Nhập số điện tháng sau: '))
tieu_thu = sau - trc
if tieu_thu <= 50:
    tien = tieu_thu * 1984
elif tieu_thu <= 100:
    tien = 50 * 1984 + (tieu_thu - 50) * 2050
elif tieu_thu <= 200:
    tien = 50 * 1984 + 50 * 2050 + (tieu_thu - 100) * 2380
elif tieu_thu <= 300:
    tien = 50 * 1984 + 50 * 2050 + 100 * 2380 + (tieu_thu - 200) * 2998
elif tieu_thu <= 400:
    tien = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + (tieu_thu - 300) * 3350
elif tieu_thu > 400:
    tien = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + 100 * 3350 + (tieu_thu - 400) * 3460
print(f'Họ và tên chủ hộ: {name}\n Tiền điện phải trả: {tien + tien * 0.08:.0f} VND')
```
